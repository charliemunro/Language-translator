```python
import collections
import numpy as np
from numpy import asarray
from numpy import savetxt
import matplotlib.pylab as plt

lis = []
lis1 = []
prob = {}
prob1 = {}
eng = 1
ita = 1

#file redircet ********

#italian_training.txt

with open('sample_words.txt', 'r') as file:
    data_eng = file.read().rstrip()
    data_eng = data_eng.lower()
    
with open('sample_words_ita.txt', 'r') as file:
    data_ita = file.read().rstrip()
    data_ita = data_ita.lower()
        
#SANTISE

chars_to_remove = [';', '"', '-','.',' ','\n','\ufeff',',']
data_eng=data_eng.translate({ord(x): '' for x in chars_to_remove})

chars_to_remove = [';', '"', '-','.',' ',',',"'",'\n']
data_ita=data_ita.translate({ord(x): '' for x in chars_to_remove})

#LIST

for i in data_eng:
    lis.append(i)
    lis = sorted(lis)

for i in data_ita:
    lis1.append(i)
    lis1 = sorted(lis1)

#COUNT

counter_eng = collections.Counter(lis)
counter_ita = collections.Counter(lis1)

#PROB

for k, v in counter_eng.items():
    prob[k] = v / sum(counter_eng.values()) 

for k, v in counter_ita.items():
    prob1[k] = v / sum(counter_ita.values())
    
#ARRAY

result = prob.items()
result1 = prob1.items()

data = list(result)
data1 = list(result1)

eng_array = np.array(data)
ita_array = np.array(data1)

#WRITE TO FILE

savetxt('eng_data.csv', eng_array, delimiter=',', fmt="%s") #fmt="%s" for string and int
savetxt('ita_data.csv', ita_array, delimiter=',', fmt="%s")

#GRAPH

keys = (prob).keys()
values = (prob).values()
plt.bar(keys, values)
plt.xlabel("Character")
plt.ylabel("Probability")
plt.title("Probability distribution for English text:")
plt.show()

keys = (prob1).keys()
values = prob1.values()
plt.xlabel("Character")
plt.ylabel("Probability")
plt.bar(keys, values)
plt.title("Probability distribution for Italian text:")
plt.show()
```
